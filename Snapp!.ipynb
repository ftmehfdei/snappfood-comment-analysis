```python
file_path = r'C:\Users\ASUS\snappfood_clean.csv'
df = pd.read_csv(file_path, encoding='cp1256')

def fix_encoding_mess(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    fixes = [] 

    try:
        fixed1 = text.encode('latin1').decode('utf-8')
        fixes.append(fixed1)
    except:
        pass  
    
    try:
        fixed2 = text.encode('cp1256').decode('utf-8')
        fixes.append(fixed2)
    except:
        pass
    
    try:
        fixed3 = text.encode('latin1').decode('cp1256').encode('latin1').decode('utf-8')
        fixes.append(fixed3)
    except:
        pass
        
    try:
        fixed4 = text.encode('windows-1256').decode('utf-8')
        fixes.append(fixed4)
    except:
        pass
    
    
    best_text = text  
    max_persian_chars = 0  
    
    for fixed in fixes:
        persian_chars = len(re.findall(r'[\u0600-\u06FF]', fixed))
        
        if persian_chars > max_persian_chars:
            max_persian_chars = persian_chars
            best_text = fixed
    
    return best_text

df['fixed_comment'] = df['clean_comment'].apply(fix_encoding_mess)
```

```python
warnings.filterwarnings('ignore')

persian_stopwords = [
    'و', 'در', 'به', 'از', 'که', 'این', 'آن', 'با', 'برای', 'تا',
    'را', 'است', 'نیست', 'ها', 'های', 'شد', 'شده', 'می', 'شود',
    'شوند', 'کرد', 'کنند', 'اما', 'اگر', 'یا', 'یک', 'هر', 'هم',
    'بود', 'هست', 'دارد', 'کردم', 'کردی', 'کند', 'بین', 'مثل',
    'قبل', 'بعد', 'وقتی', 'زمانی', 'هیچ', 'چند', 'حتی', 'خیلی'
]

def preprocess_persian_text(text):
    if pd.isna(text):
        return ""
    
    text = str(text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[^\u0600-\u06FF\s]', ' ', text)
    text = ' '.join(text.split()) 
    return text

def remove_stopwords(text):
    if pd.isna(text) or text == "":
        return ""
    
    words = text.split()
    words = [w for w in words if w not in persian_stopwords and len(w) > 1]
    return ' '.join(words)

df['cleaned_nlp'] = df['fixed_comment'].apply(preprocess_persian_text)
df['cleaned_no_stopwords'] = df['cleaned_nlp'].apply(remove_stopwords)

from IPython.display import display

display_df = sample_df[[
    'label', 
    'fixed_comment', 
    'cleaned_nlp', 
    'cleaned_no_stopwords',
]]

pd.set_option('display.max_colwidth', 200)  
pd.set_option('display.max_columns', None)

display(display_df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def get_top_ngrams(text_series, ngram_range=(2,2), n=15):

    vectorizer = CountVectorizer(ngram_range=ngram_range).fit(text_series)
    bag_of_words = vectorizer.transform(text_series)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
    return words_freq[:n]

print("Top Words")
top_bigrams = get_top_ngrams(df['cleaned_nlp'].dropna())
for bigram, count in top_bigrams:
    print(f"{bigram}: {count}")

print("\nHappy Comments")
top_bigrams_happy = get_top_ngrams(df[df['label']=='HAPPY']['cleaned_nlp'].dropna())
for bigram, count in top_bigrams_happy:
    print(f"{bigram}: {count}")

print("\nSad Comments")
top_bigrams_sad = get_top_ngrams(df[df['label']=='SAD']['cleaned_nlp'].dropna())
for bigram, count in top_bigrams_sad:
    print(f"{bigram}: {count}")
```
